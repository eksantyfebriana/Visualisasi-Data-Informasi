## **Data yang digunakan**

Dari dataset monkeypox.csv atribut yang dipilih untuk visualisasi ini yaitu:

*   location: Nama lokasi
*   date: Tanggal data dikumpulkan
*   total_cases: Jumlah total kasus yang dikonfirmasi
*   total_deaths: Jumlah total kematian yang dikonfirmasi.
*   new_cases: Kasus baru per hari.
*   new_deaths: Kematian baru per hari.

## **Tujuan, User, Fungsi, dan Tone**

**a. Tujuan**:
*   Menampilkan perkembangan kasus dan kematian monkeypox dari waktu ke waktu berdasarkan lokasi.
*   Mempermudah pemahaman tren epidemiologis penyakit monkeypox di berbagai wilayah, termasuk tren kasus baru dan kematian baru.
</n>

**b. User**:
*   Ahli epidemiologi, peneliti kesehatan, dan pembuat kebijakan yang tertarik dalam memantau penyebaran penyakit.
*   Publik umum atau pejabat kesehatan masyarakat yang ingin mendapatkan wawasan tentang seberapa serius penyebaran penyakit di lokasi tertentu.
</n>

**c. Fungsi**:
*   Memberikan representasi visual yang informatif tentang tren kasus dan kematian untuk memfasilitasi pengambilan keputusan dan pemantauan situasi kesehatan.
*   Membantu dalam memahami lonjakan kasus atau penurunan serta efektivitas intervensi yang mungkin dilakukan di setiap lokasi.
</n>

**d. Tone**:
*   Serius, profesional, dan informatif, dengan fokus pada penyampaian data kesehatan publik secara jelas dan tepat.
*   Menggunakan visualisasi yang sederhana namun kuat untuk menyampaikan tren yang penting.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = 'monkeypox.csv'
monkeypox_data = pd.read_csv(file_path)

# Convert 'date' to datetime format
monkeypox_data['date'] = pd.to_datetime(monkeypox_data['date'])

# Set up the Seaborn style
sns.set(style="whitegrid")

# 1. Visualisasi Kasus Baru dan Kematian Baru
plt.figure(figsize=(14, 8))

# Plot New Cases
sns.lineplot(data=monkeypox_data, x='date', y='new_cases', hue='location', marker='o', linewidth=2)
plt.title('New Monkeypox Cases Over Time by Location', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('New Cases', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Plot New Deaths
plt.figure(figsize=(14, 8))
sns.lineplot(data=monkeypox_data, x='date', y='new_deaths', hue='location', marker='o', linewidth=2)
plt.title('New Monkeypox Deaths Over Time by Location', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('New Deaths', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 2. Visualisasi Total Kasus dan Kematian
plt.figure(figsize=(14, 8))

# Plot Total Cases
sns.lineplot(data=monkeypox_data, x='date', y='total_cases', hue='location', marker='o', linewidth=2)
plt.title('Total Monkeypox Cases Over Time by Location', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Cases', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Plot Total Deaths
plt.figure(figsize=(14, 8))
sns.lineplot(data=monkeypox_data, x='date', y='total_deaths', hue='location', marker='o', linewidth=2)
plt.title('Total Monkeypox Deaths Over Time by Location', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Deaths', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
